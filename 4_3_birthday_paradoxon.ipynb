# Birthday Paradoxon

```python
# Import packages
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

```python
# Initialize variables
d = 365 # Days per year
max_people = 70
math_results = np.ones(max_people) # Initialize all probabilities to 1
```

```python
# Calculate probabilities
for p in range(max_people):
    
  # Calculation of series: 365/365 * 364/365 * ... * (365 - p + 1)/365
  for i in range(p):
    math_results[p] = math_results[p] * (d - i) / d

  # Result is inverse probability
  math_results[p] = (1 - math_results[p]) * 100
```

```python
# Visualize results
plt.title("Geburtstagsparadoxon (Modell)")
plt.xlabel("Anzahl Personen")
plt.ylabel("Wahrscheinlichkeit")
plt.minorticks_on()
plt.grid()
plt.grid(which="minor", linestyle=":")
plt.plot(math_results)
```

```python
# Initialize variables
max_people = 1
experiments = 100
math_results = np.ones(experiments)
```

```python
# Conduct experiment
for l in range(0, experiments):
  results = np.zeros(max_people)
  
  # Create random birthdates and store them in a vector
  for i in range(0, max_people):
    bday = rd.randint(1, 365)
    results[i] = bday
    
  # Compare birthdates
  same = 0
  for k in range(max_people-1):
    for j in range(k+1, max_people):
      if results[k] == results[j]:
        same = same + 1
        
  # Check if there is at least one pair
  if same == 0:
    math_results[l] = 0
  
  # Print number of same birthdates
  print("Anzahl Treffer: ", same)

  # Update counter
  max_people = max_people + 1
```

```python
# Output
print(math_results)
```

```python
# Visualize results
plt.clf()
plt.title("Geburtstagsparadoxon")
plt.xlabel("Anzahl Personen")
plt.ylabel("Wahrscheinlichkeit")
plt.minorticks_on()
plt.grid()
plt.grid(which="minor", linestyle=":")
plt.plot(results, label="Monte-Carlo-Simulation")
plt.plot(math_results, label="Mathematisches Modell")
plt.legend(loc='upper left')
```
