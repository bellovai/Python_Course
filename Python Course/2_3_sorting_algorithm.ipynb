```python
# Find maximum
maximum = int(0)
for i in range(0, 122):
  if maximum < pollen[i]:
    maximum = pollen[i]
    savevar = i
```

```python
# Output results
print("Maximum: ", maximum)
print("Datum: ", messzeitpunkte[savevar])
```

```python
# Define algorithm: bubble sort
z = True
while z == True:
  z = False
  for i in range(0,121):
    if pollen[i] > pollen[i+1]:
      savevar2 = pollen[i]
      pollen[i] = pollen[i+1]
      pollen[i+1] = savevar2
      savevar3 = messzeitpunkte[i]
      messzeitpunkte[i] = messzeitpunkte[i+1]
      messzeitpunkte[i+1] = savevar3
      z = True
```
