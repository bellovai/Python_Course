# Biological Model

```python
# Import packages
import matplotlib.pyplot as plt
```

```python
# Initialize variables
k = 50 # Capacity
v = 0.5 # Rate of multiplication
l = 30 # Length of simulation
t = [0 for x in range(l)] # Time list
N = [0 for x in range(l)] # Number of lists
N[0] = 2.0
```

```python
# Define algorithm
for i in range(1, l):
  N[i] = round(N[i-1] + v*N[i-1]*((k-N[i-1])/k), 3)
for i in range(1, l):
  t[i] = i
```

```python
# Output results
plt.plot(N)
plt.title("Anzahl Kaninchen über die Zeit")
plt.xlabel("Zeit")
plt.ylabel("Anzahl Kaninchen")
```
