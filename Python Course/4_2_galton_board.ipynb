# Galton Board

```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Input number of levels and balls
höhe = int(input("Höhe? "))
kugeln = int(input("Anzahl Kugeln? "))
```

```python
# Create vector to track location
behaelter = np.array([0]*(höhe+1))
```

```python
# Create tracker of location
for j in range(0, kugeln):
  print()
  print("Kugel ", j+1)
  summe = 0
  for i in range(0, höhe):
    # Create random number between 0 and 1
    prob = random.uniform(0, 1)
    
    # Transform probability to bool
    decision = 1
    if prob < 0.5:
      decision = 0
    
    # Update tracker of location  
    summe = summe + decision
    
    # Print bool
    print(decision)
  
  behaelter[summe] = behaelter[summe] + 1
```

```python
# Print tracker of location
print()
print("Kugeln in den Behaeltern:")
print(behaelter)
```

```python
# Visualize results
plt.bar(range(höhe+1), behaelter)
plt.title("Galton Board")
plt.xlabel("Behälter")
plt.ylabel("Anzahl Kugeln")
```
